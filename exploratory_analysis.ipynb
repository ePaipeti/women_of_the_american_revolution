### Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Load corpus and remove duplicates

```python
# load clean corpus
df = pd.read_csv("clean_corpus.csv")
df.info()
```

```python
# remove duplicates and inspect the dataframe
df.drop_duplicates(subset=['text'], inplace = True)
df.info()
```

Create gendered corpora

```python
# retrieve letters written by men
men_df = df[df['gender'] == 'male']
```

```python
# retrieve letters written and addressed to men
men_to_men = men_df[men_df['recipient_gender'] == 'male']
```

```python
men_to_men.info()
```

```python
# retrieve letters by men authors addressed to women
men_to_women = men_df[men_df['recipient_gender'] == 'female']
men_to_women.info()
```

```python
# retrieve letters written by women
women_df = df[df['gender'] == 'female']
women_df.info()
```

```python
# retrieve letters written and addressed to women
women_to_women = women_df[women_df['recipient_gender'] == 'female']
women_to_women.info()
```

```python
# retrieve letters by women authors addressed to men
women_to_men = women_df[women_df['recipient_gender'] == 'male']
women_to_men.info()
```

### Visualize letter distribution over the years

```python
# Step 1: Convert 'date_from' to datetime
df['date_from'] = pd.to_datetime(df['date_from'])

# Step 2: Extract the year from 'date_from'
df['year'] = df['date_from'].dt.year

# Step 3: Count the number of letters written each year
yearly_counts = df['year'].value_counts().sort_index()

# Plotting the results
plt.figure(figsize=(12, 6))
yearly_counts.plot(kind='bar')
plt.xlabel('Year')
plt.ylabel('Number of Letters')
plt.title('Number of Letters Written Each Year')

# Rotate x-axis labels by 45 degrees and set font size to 8
plt.xticks(rotation=45, fontsize=8)

# Reduce the number of x-ticks
plt.locator_params(axis='x', nbins=20)

plt.show()
```

```python
revolutionary_letters = df[df['period'] == 'Revolutionary War']
```

```python
revolutionary_letters.groupby('gender').count()
```

```python
revolution_women = revolutionary_letters[revolutionary_letters['gender'] == 'female']
```

```python
unique_women_auth_r = revolution_women['authors'].unique()
unique_women_auth_r 
```

```python
# Group by 'source', count occurrences, and sort values
source_counts = revolution_women.groupby('authors').count().sort_values(by='title', ascending=False)


# Filter out sources with only one occurrence
source_counts_filtered = source_counts[source_counts['title'] > 5]

# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(x=source_counts_filtered.index, y=source_counts_filtered['title'], palette='viridis')
plt.xlabel('Authors')
plt.ylabel('Count')
plt.title('Count of Occurrences by Author (Revolutionary War) (Excluding Single Occurrences)')
plt.xticks(rotation=45, ha='right', fontsize= 7)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

```python
counts = women_to_women.groupby('authors').count().sort_values(by= 'text', ascending = False)

# Filter out sources with only one occurrence
counts_filtered = counts[counts['text'] > 3]

# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(x=counts_filtered.index, y=counts_filtered['text'], palette='viridis')
plt.xlabel('Authors')
plt.ylabel('Count')
plt.title('Count of Occurrences by Author (All Periods) (Excluding Single Occurrences)')
plt.xticks(rotation=45, ha='right', fontsize= 7)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```
